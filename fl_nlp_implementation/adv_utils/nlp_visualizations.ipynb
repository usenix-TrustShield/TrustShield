```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
arfed = []
vfl = []
tshield = []
tshield_g = []
ratios = np.round(np.linspace(0,1,21),2)
for ratio in ratios:
        vfl.append(np.load("../histories/NLP/FLNLP_mining_0_noniidedge_0_uniform_ratio_"+str(ratio)+".npy"))
        tshield.append(np.load("../histories/NLP/FLNLP_mining_1_noniidedge_0_uniform_ratio_"+str(ratio)+".npy"))
        tshield_g.append(np.load("../histories/NLP/FLNLP_mining_1_noniidedge_0_gaussian_ratio_"+str(ratio)+".npy"))
        arfed.append(np.load("../histories/NLP/FLNLP_ARFED_isnoniid_0_"+str(ratio)+".npy"))
vfl = np.array(vfl)
arfed = np.array(arfed)
tshield = np.array(tshield)
tshield_g = np.array(tshield_g)
```

```python
plt.plot(np.arange(0,105,5),vfl[:,-1],'-o')
plt.plot(np.arange(0,105,5),tshield[:,-1], '-o')
plt.plot(np.arange(0,105,5),tshield_g[:,-1], '-o')
plt.plot(np.arange(0,105,5),arfed[:,-1], '-o')
plt.legend(["Vanilla FL", "TrustShield", "TrustShield G", "ARFED"])
```
