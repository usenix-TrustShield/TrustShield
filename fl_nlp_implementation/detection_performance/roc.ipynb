```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
hostile_nodes = [0,1,2,6, 8, 9, 11, 12, 13, 14, 15, 16, 17, 19]
collaborative_nodes = [3, 4, 5, 7, 10, 18]

filtered = [['model3', 'model4', 'model5', 'model7', 'model10', 'model18'], ['model0', 'model1', 'model2', 'model3', 'model4', 'model5', 'model6', 'model7', 'model8', 'model9', 'model10', 'model11', 'model13', 'model14', 'model16', 'model17', 'model18', 'model19'], ['model0', 'model1', 'model3', 'model4', 'model5', 'model7', 'model8', 'model9', 'model10', 'model11', 'model12', 'model13', 'model14', 'model17', 'model18', 'model19'], ['model0', 'model1', 'model2', 'model3', 'model4', 'model5', 'model7', 'model8', 'model9', 'model10', 'model11', 'model12', 'model13', 'model14', 'model16', 'model17', 'model18', 'model19'], ['model1', 'model3', 'model4', 'model5', 'model7', 'model8', 'model9', 'model10', 'model11', 'model13', 'model16', 'model17', 'model18', 'model19'], ['model3', 'model4', 'model5', 'model7', 'model9', 'model10', 'model11', 'model14', 'model17', 'model18'], ['model0', 'model3', 'model4', 'model5', 'model7', 'model8', 'model9', 'model10', 'model11', 'model16', 'model17', 'model18'], ['model0', 'model1', 'model3', 'model4', 'model5', 'model7', 'model10', 'model11', 'model16', 'model17', 'model18', 'model19'], ['model0', 'model1', 'model3', 'model4', 'model5', 'model7', 'model10', 'model11', 'model16', 'model17', 'model18', 'model19'], ['model3', 'model4', 'model5', 'model7', 'model10', 'model11', 'model16', 'model18'], ['model3', 'model4', 'model5', 'model7', 'model10', 'model11', 'model18'], ['model3', 'model4', 'model5', 'model7', 'model10', 'model11', 'model18'], ['model3', 'model4', 'model5', 'model7', 'model10', 'model18'], ['model0', 'model1', 'model2', 'model3', 'model4', 'model5', 'model6', 'model7', 'model8', 'model10', 'model13', 'model14', 'model15', 'model18'], ['model3', 'model4', 'model5', 'model7', 'model10', 'model18']]
filtered_in_nums = []
for roundd in range(len(filtered)):
    round_array = []
    for elem in filtered[roundd]:
        round_array.append(int(elem[5:]))
    filtered_in_nums.append(round_array)

true_detection = []
false_detection = []
initialization_rounds=5
for roundd in range(initialization_rounds):
    true_detection.append(1)
    false_detection.append(1)
for roundd in filtered_in_nums:
    true_detection.append(np.sum(np.isin(np.asarray(roundd), collaborative_nodes))/len(collaborative_nodes))
    false_detection.append(np.sum(np.isin(np.asarray(roundd), hostile_nodes))/len(hostile_nodes))
```

```python
plt.plot(np.arange(1,len(true_detection)+1,1), true_detection, '-o')
plt.plot(np.arange(1,len(true_detection)+1,1),false_detection, '-o')
```
