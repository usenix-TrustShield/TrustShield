```python
plt.plot(np.linspace(0,100,21).round(2),np.transpose(accuracies_non_noniid)[-1],'-o',lw=2)
plt.plot(np.linspace(0,100,21).round(2),np.transpose(accuracies_uni_noniid)[-1],':>', mfc='none',lw=2)
plt.plot(np.linspace(0,100,21).round(2),np.transpose(accuracies_gaus_noniid)[-1],':<', mfc='none',lw=2)
plt.plot(np.linspace(0,100,21).round(2),np.transpose(accuracies_ARFED_noniid)[-1],':s', mfc='none',lw=2)

plt.xlabel("Adversarial User Ratio")
plt.ylabel("Accuracy")
plt.title("MNIST FL Performance for Different Settings (Non-IID)")
plt.legend(["No Prevention", "Uniform Validators", "Gaussian Validators", "ARFED"])
plt.xlim([0,100])
plt.ylim([0,1])
plt.grid(True)
```
