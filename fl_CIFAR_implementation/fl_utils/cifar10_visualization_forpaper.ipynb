```python
ratios = np.linspace(0,1,21)
histories = []
uniform_histories = []
gaussian_histories = []
ARFED_histories = []
for ratio in ratios:
    histories.append(np.load("../histories/CIFAR10/FLCIFAR10_mining_0_noniidedge_0_uniform_ratio_"+str(round(ratio,2))+".npy"))
    uniform_histories.append(np.load("../histories/CIFAR10/FLCIFAR10_mining_1_noniidedge_0_uniform_ratio_"+str(round(ratio,2))+".npy"))
    gaussian_histories.append(np.load("../histories/CIFAR10/FLCIFAR10_mining_1_noniidedge_0_gaussian_ratio_"+str(round(ratio,2))+".npy"))
    ARFED_histories.append(np.load("../histories/CIFAR10/FLCIFAR10_ARFED_isnoniid_0_"+str(round(ratio,2))+".npy"))
#ARFED_histories = np.array(ARFED_histories)
histories = np.array(histories)
uniform_histories = np.array(uniform_histories)
gaussian_histories = np.array(gaussian_histories)
ARFED_histories = np.array(ARFED_histories)
```
